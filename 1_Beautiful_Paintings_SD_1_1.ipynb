```python
import torch
from torch import autocast
from diffusers import StableDiffusionPipeline

model_id = "CompVis/stable-diffusion-v1-1"
device = "cuda"

pipe = StableDiffusionPipeline.from_pretrained(model_id)
pipe = pipe.to(device)
```

```python
# Set the random seed for reproducibility
torch.manual_seed(56)

prompt = "a beautiful painting of a man and a woman"
results = pipe(prompt, guidance_scale=7.5, num_inference_steps=50)

image = results.images[0]

image.save("1 sd 1.1 04.png")
image
```

```python
import torch
import matplotlib.pyplot as plt
from torch.cuda.amp import autocast

# Define the prompt
prompt = "a beautiful painting of a man and a woman"

# Generate 24 images with different seeds
images = []
for seed in range(42, 42+24):
    torch.manual_seed(seed)
    results = pipe(prompt, guidance_scale=7.5, num_inference_steps=50)
    images.append(results.images[0])

images[0]

# Display images in an 8x3 grid
fig, axes = plt.subplots(nrows=3, ncols=8, figsize=(16, 6))
for ax, img in zip(axes.flatten(), images):
    ax.imshow(img)
    ax.axis("off")

plt.tight_layout()
plt.show()
```
